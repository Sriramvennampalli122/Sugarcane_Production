using distplot to understand the type of distribution happening,Production (Tons),Production per Person (Kg),Acreage (Hectare),Yield (Kg / Hectare)

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(df["Production (Tons)"])
plt.subplot(2,2,2)
sns.distplot(df["Production per Person (Kg)"])
plt.subplot(2,2,3)
sns.distplot(df["Acreage (Hectare)"])
plt.subplot(2,2,4)
sns.distplot(df["Yield (Kg / Hectare)"])
```

From the distribution graph of production(Tons) column we can observe that the bell curve is from 0 to 2 on x-axis which means mostly of the countries produce (0to2)x10^8 tons at most. There is an outlier at 0 which indicates large number of countries produce less than 1x10^8 ton.we can also observe that the graph is right skewed which indicate there were countries which produce more than 200 million tons

From the distribution graph of production per person(kg) column we can observe that the bell curve is froming at -1 to 2 on x-axis which means in most of the countries a preson can produce less than 10^6 kgs to 2x10^6 kgs at most. There is an outlier at 0 which indicates in large number of countries production per person is less than 10^8 kgs.we can also observe that the graph is right skewed which indicate there were countries in which production per person is more than 2 million kgs

From the distribution graph of the Acreage (Hectare) column, we can observe that the bell curve spans from 0 to 0.2 on the x-axis. This indicates that most sugarcane-producing countries have less than 1 million hectares or, at most, 2 million hectares of land dedicated to sugarcane cultivation. There is an outlier at 0, which suggests that a significant number of countries have close to 1 million hectares of land for sugarcane production. Additionally, the graph is right-skewed, indicating the presence of some countries with more than 10 million hectares of land allocated for sugarcane cultivation.

From the distribution graph of the Yield (kg/Hectare) column, we can observe that the data does not form a perfect bell curve. It can be noted that the majority of countries have a yield ranging between 1 to 1.5 million kg per hectare, with the highest concentration of countries achieving yields of more than 1 million kg per hectare.

Grouping dataset using continent attribute and creating a new dataset

```python
df_cont=df.groupby("Continent").sum()
```

```python
df_cont
```

```python
df_cont=df_cont.sort_values("Production (Tons)",ascending=False)
```

Arranging the continents with the number of countries in them

```python
df["Continent"].value_counts()
```

```python
df_cont["Production (Tons)"].plot(kind="bar")
```

From the above bar graph, it can be observed that South America has the highest sugarcane production, followed by Asia and North America. This indicates that sugarcane production is not directly related to the number of countries within a continent, as some continents with fewer countries still have higher production due to favorable conditions and resources.

```python
df_cont["Acreage (Hectare)"].plot(kind="bar")
```

From the above bar graph, it can be observed that South America has the highest sugarcane production, followed by Asia and North America. This suggests that sugarcane production is directly linked to the cultivation area available within each continent, highlighting the importance of large-scale agricultural land in driving higher production volumes.

```python
df_cont["no_of_countries"]=df.groupby("Continent").count()["Country"]
```

```python
df_cont
```

```python
cont_nam=df_cont.index.to_list()
sns.lineplot(data=df_cont,x="no_of_countries",y="Production (Tons)")
plt.xticks(df_cont["no_of_countries"],cont_nam,rotation=90)
plt.show()
```

The above line graph illustrates the relationship between sugarcane production and the number of countries in each continent. It highlights how production levels vary regardless of the number of countries, providing insights into regional contributions to global sugarcane output.
